```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Create results folder
os.makedirs("results", exist_ok=True)

# Models
models = ['Random Forest', 'SVR', 'XGBoost']
x = np.arange(len(models))
width = 0.25

# -----------------------------
# Metrics from your training output
# -----------------------------

# RMSE values
chl_rmse = [1.6716, 3.1122, 1.6158]
do_rmse  = [2.1757, 3.3990, 2.2242]
nh3_rmse = [0.6951, 0.9995, 0.5671]

# R² values
chl_r2 = [0.8892, 0.6161, 0.8965]
do_r2  = [0.7895, 0.4863, 0.7800]
nh3_r2 = [0.9525, 0.9019, 0.9684]

# MAPE values
chl_mape = [4.1557, 6.1500, 4.0589]
do_mape  = [21.6847, 26.6541, 21.7180]
nh3_mape = [5.4849, 5.5293, 4.9707]

# -----------------------------
# Function to create grouped bar plot
# -----------------------------
def plot_grouped(metric1, metric2, metric3, title, ylabel, filename):
    plt.figure(figsize=(8,5))

    plt.bar(x - width, metric1, width, label='Chlorophyll-a')
    plt.bar(x, metric2, width, label='Dissolved Oxygen')
    plt.bar(x + width, metric3, width, label='NH3-N')

    plt.xlabel("Models")
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(x, models)
    plt.legend()

    plt.savefig(f"results/{filename}", dpi=300)
    plt.close()

# -----------------------------
# Generate combined plots
# -----------------------------

print("Generating combined model comparison plots...")

plot_grouped(chl_rmse, do_rmse, nh3_rmse,
             "RMSE Comparison Across Models",
             "RMSE",
             "combined_rmse.png")

plot_grouped(chl_r2, do_r2, nh3_r2,
             "R² Comparison Across Models",
             "R² Score",
             "combined_r2.png")

plot_grouped(chl_mape, do_mape, nh3_mape,
             "MAPE Comparison Across Models",
             "MAPE (%)",
             "combined_mape.png")

print("Plots saved in 'results/' folder")
```
